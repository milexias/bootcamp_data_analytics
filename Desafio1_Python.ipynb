2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões?` [‘mean’, np.median, pd.Series.mod]`

```python
media_sono_profissao = dados.groupby('Profissão')['Duração do sono'].mean().round(2).reset_index()
media_sono_profissao
```

```python
mediana_sono_profissao = dados.groupby('Profissão')['Duração do sono'].median().reset_index()
mediana_sono_profissao
```

```python
moda_sono_profissao = dados.groupby('Profissão')['Duração do sono'].apply(lambda x:
                                                                          x.mode().iloc[0]).reset_index()

moda_sono_profissao
```

3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?

```python
eng_software = dados['Profissão'].value_counts()['Eng. de Software']

eng_software_obesos = dados[(dados['Profissão']=='Eng. de Software') & (dados['Categoria IMC']=='Obesidade')]['Categoria IMC'].count()

print(f'Das pessoas que atuam com Engenheiros de Software, {eng_software_obesos / eng_software * 100:.0f}% sao obesos.')
```

9. É predominante entre os participantes dormir 8 horas por dia?
`(considere usar Moda como medida)`

```python
moda_duracao_sono = dados['Duração do sono'].mode()

if moda_duracao_sono.values >= 8:
    print("Dormir 8 horas por dia é predominante.")
else:
    print("Dormir 8 horas por dia não é predominante.")
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? `(Use a média)`

```python
passos_acima_70 = dados[dados['Frequência cardíaca'] > 70]['Passos diários'].mean()
passos_70_menos = dados[dados['Frequência cardíaca'] <= 70]['Passos diários'].mean()

if passos_acima_70 > passos_70_menos:
    print("Pessoas com frequências cardíacas acima de 70 dão mais passos.")
else:
    print("Pessoas com frequências cardíacas de 70 ou menos, dão mais passos ou possuem a mesma média de passos.")
```
